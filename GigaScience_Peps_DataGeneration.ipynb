````python
# Helper function to create new config file

import json
import os

def newConfigFile(outputPrefix,inputPrefix,outfile,v1,v2,v3,v4,v5):
    with open("PEPS/SampleData/config-small.json", 'r') as f:
        config = json.load(f)

        config['inputType']='csv'
        config['outputPrefix']=outputPrefix
        config['inputPrefix']=inputPrefix
        config['variables'][0]['numVar']=v1
        config['variables'][1]['numVar']=v2
        config['variables'][2]['numVar']=v3
        config['variables'][3]['numVar']=v4
        config['variables'][4]['numVar']=v5

    os.makedirs(os.path.dirname(outfile), exist_ok=True)
    with open(outfile+".json", 'w') as outfile:
        json.dump(config, outfile, sort_keys=True, indent=4)

        
````

````python
#helper function to train dataset

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score

def RF_AUC(dfx, nTree):
    df = dfx.copy()
    features = df.columns[:-1]
    
    df['is_train'] = np.random.uniform(0, 1, len(df)) <= .75
    train, test = df[df['is_train']==True], df[df['is_train']==False]
    clf = RandomForestClassifier(n_jobs=2, n_estimators=nTree, random_state=0)
    clf.fit(train[features], train['lbl'])
    #clf.predict(test[features])
    prob = clf.predict_proba(test[features])
    y_true = test['lbl']
    y_scores = prob[:,1]
    return clf, roc_auc_score(y_true, y_scores)
````

````python
from scipy.stats import sem
from statistics import mean 

numTree=500

AucTruthAr=[]
AucTruthAr100=[]

# run 5 times to get a feel for the variation
for i in range(0,5):
    clfTruth, AucTruth = RF_AUC(jnAll, nTree=numTree)
    AucTruthAr.append(AucTruth)
    clfTruth, AucTruth = RF_AUC(jn100, nTree=numTree)
    AucTruthAr100.append(AucTruth)

print("AUC Total                    : ", mean(AucTruthAr),"sem", sem(AucTruthAr) )
print("AUC VariantSpark 100         : ", mean(AucTruthAr100),"sem", sem(AucTruthAr100) )
````
